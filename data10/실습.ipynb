```python
df_summary_top100 = df_summary[:100].copy()
print(df_summary_top100.shape)
df_summary_top100.head()
```

```python
df_summary_top100.drop(columns='Standard error of ladder score', inplace=True)
df_summary_top100.drop(columns='upperwhisker', inplace=True)
df_summary_top100.drop(columns='lowerwhisker', inplace=True)
df_summary_top100.drop(columns='Social support', inplace=True)
df_summary_top100.head()
```

```python
pivot_table = df_summary_top100.pivot_table(index=['Regional indicator'],
                                            values=['Ladder score','Healthy life expectancy'],
                                            aggfunc=[len,np.mean])
pivot_table
```

```python
pivot_table.columns = ['Healthy life expectancy_len','Ledder score_len','Healthy life expectancy_mean','Ledder score_mean']
pivot_table
```

```python
f, ax = plt.subplots(3,1,figsize=(8,25))
pivot_table['Ledder score_len'].plot.pie(explode=[0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02], ax=ax[0], autopct='%2.1f%%')
pivot_table['Ledder score_mean'].plot.barh(ax=ax[1])
pivot_table['Healthy life expectancy_mean'].plot.barh(ax=ax[2])
ax[0].set_title('행복지수 top100중 대륙별 숫자')
ax[1].set_title('각 대륙별 행복지수 평균')
ax[2].set_title('각 대륙별 기대수명 평균')
plt.show()
```
