2. 항목들이 크게 분석할 것과 상관이 없으면 제거  
    1) 수학적 공분산을 이용해 계산 -> 숫자가 커지면 공분산이 커진다  
    2) 상관도 계산 -> 숫자를 -1~1사이의 범위로 정해서 상관관계를 계산  
  
Feature들을 선택, 전처리: feature engineering(속성 공학)

```python
# 각 항목들의 상관도를 구해줌(-1~1)
raw_data.corr()
# 상관 관계 결과
# 0: 컬럼간에 전혀 상관이 없다
# 1: 완벽히 상관이 있다(양)
# -1: 환벽히 상관이 있다(음)
# (+-)0.3: 약간의 상관이 있다
# (+-)0.5: 상관이 있는거 같다
# (+-)0.7: 꾀 큰 상관이 있다
```

```python
# 앞서 Miss를 구한 방식을 lambda에 적용해서 conversion_rare에 저장
conversion_rare = lambda x: x.split(',')[1].split('.')[0].strip()
# name 컬럼의 값을 하나씩 거치면서 분리되어 나온 결과를 title이라는 컬럼을 생성해 저장
raw_data['title'] = raw_data['name'].map(conversion_rare)
# 나온 결과의 모든 값들을 하나씩만 출력
titles = raw_data['title'].unique()
titles
```

```python
# title 컬럼의 각 값에 대해서 성별 인원수를 출력
pd.crosstab(raw_data['title'], raw_data['sex'])
```

```python
raw_data['title'] = raw_data['title'].replace('Mlle','Miss')# Mlle를 Miss로 변환
raw_data['title'] = raw_data['title'].replace('Ms','Miss')# Ms를 Miss로 변환
raw_data['title'] = raw_data['title'].replace('Mme','Mrs')# Mme를 Mrs로 변환

# 드물게 표현되는 값에 대해서 하나로 통합해서 처리해버림
Rare = ['Col','Dr', 'Major', 'Capt', 'Lady', 'Sir', 'Dona', 'Master', 'Jonkheer', 'the Countess', 'Don', 'Rev']
for x in Rare:
    # 드물게 표현된 모든 값을 Rare로 변환
    raw_data['title'] = raw_data['title'].replace(x,'Rare')
    
raw_data['title'].unique()
```

```python
# 'title'의 각 값에 해당하는 생존자들의 평균
raw_data[['title','survived']].groupby(['title']).mean()
```

```python
# raw_data를 깊은 복사를해서 raw_data2를 만들고
# sex 컬럼의 값을 모두 'f'로 바꿔줌
tmp = ['f']*1309
raw_data2 = raw_data.copy()
raw_data2['sex'] = tmp
raw_data2['sex']
```

```python
# 비어있는 tmp2 라는 배열을 생성하고,
# raw_data의 sex컬럼의 값들일 비교해가면서  femail은 0, male은 1로 tmp2에 저장
tmp2 = []
for x in raw_data['sex']:
    if x == 'female':
        tmp2.append(0)
    elif x == 'male':
        tmp2.append(1)
    else:
        tmp2.append(np.nan)
len(tmp2)
```

```python
raw_data['sex'] = tmp2
raw_data.head()
```

```python
# 각 컬럼의 타입변환 to_frame, astype('float')
raw_data['survived'] = raw_data['survived'].astype('float')
raw_data['pclass'] = raw_data['pclass'].astype('float')
raw_data['sex'] = raw_data['sex'].astype('float')
raw_data['sibsp'] = raw_data['sibsp'].astype('float')
raw_data['parch'] = raw_data['parch'].astype('float')
raw_data['fare'] = raw_data['fare'].astype('float')
raw_data.head()
```

```python
# 컬럼의 NaN을 제거
raw_data = raw_data[raw_data['age'].notnull()]
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.info()
```

```python
#훈련 시킬 컬럼을
train_pre = raw_data[['pclass','sex','age','sibsp','parch','fare']]
train_pre.head()
```

# ---------------------- data 준비 끝 ----------------------

공부할 데이터/검증용 데이터 분리

```python
#머신러닝용 라이브러리: sklearn, 사이킷 럿
from sklearn.model_selection import train_test_split
```

```python
                                                #(데이터프레임, 맞춰야할 정답, 테스트데이터10#지정, seed값)
X_train, X_test, y_train, y_test = train_test_split(train_pre, raw_data[['survived']], test_size=0.1, random_state = 13)
#각 변수에 저장
```

```python
X_train = X_train.reset_index()
X_train = X_train.drop(['index'], axis=1)

X_test = X_test.reset_index()
X_test = X_test.drop(['index'], axis=1)

y_train = y_train.reset_index()
y_train = y_train.drop(['index'], axis=1)

y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
# classifier는 분류, Regressor는 예측

# 파라메터를 분석가가 임의로 변결할 수 없음
# 파라메터를 분서가가가 변경 가능, 변경 불가능
# 분석가가 변경 가능한 파라메터를 '하이퍼 파라미터라함'
# 학습기 객체 생성
tree_clf = DecisionTreeClassifier(max_depth = 3, random_state=13)
# 공부 시키기
tree_clf.fit(X_train,y_train)

# train 데이터를 가지고 공부한 점수
print('Score: {}'. format(tree_clf.score(X_train, y_train)))
```

```python
from sklearn.metrics import accuracy_score

y_pred = tree_clf.predict(X_test)# 문제를 주고 플어보라고 함

# 테스트 데이터를 푼 답안과 해답을 비교홰 정확도를 구해보자
accuracy_score(y_test, y_pred)
```

```python
def isSurvived(name, person):
    isSurvive = 'not survived' if tree_clf.predict([person])[0] == 0 else 'survived'
    print(name, ' is ', isSurvive, 
          ' --> ', max(tree_clf.predict_proba([person])[0]))
    
isSurvived('Dicaprio', dicaprio)
isSurvived('Winslet', winslet)
```
