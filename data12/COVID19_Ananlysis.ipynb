```python
f,ax=plt.subplots(figsize=(15,5))
sns.countplot('province', data=rd_case, ax=ax)
plt.xticks(rotation=45)
```

```python
f,ax=plt.subplots(figsize=(15,7))
sns.lineplot(x='date',y='confirmed', data=rd_time, ax=ax)
plt.xticks(np.arange(0,163,20))
```

```python
f,ax=plt.subplots(figsize=(15,7))
sns.lineplot(x='date',y='test', data=rd_time, ax=ax)
plt.xticks(np.arange(0,163,20))
```

```python
f,ax=plt.subplots(figsize=(15,7))
sns.lineplot(x='date',y='negative', data=rd_time, ax=ax)
plt.xticks(np.arange(0,163,20))
```

```python
f,ax1=plt.subplots(figsize=(15,7))
ax2=ax1.twinx()
sns.lineplot(x='date',y='negative', data=rd_time, ax=ax1, color="blue",label='Negative')
plt.xticks(np.arange(0,163,20))
sns.lineplot(x='date',y='confirmed', data=rd_time, ax=ax2, color="red", label='Confirmed')
ax1.legend(loc='upper left',fontsize=20)
ax2.legend(loc='lower right',fontsize=20)
ax1.set_ylabel('Negative', fontsize=15, color='blue')
ax2.set_ylabel('Confirmed', fontsize=15, color='red')
plt.yticks(np.arange(0,60000,5000))
plt.xticks(np.arange(0,163,20))
```
